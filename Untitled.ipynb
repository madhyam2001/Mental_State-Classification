```python
import tensorflow.keras.layers as tfl
import tensorflow.random as tfr
from keras.models import Sequential
```

```python
input_shape=(988,1)
i=tfr.normal(input_shape)
model = Sequential()
model.add(tfl.Conv1D(64, 1, activation='relu', input_shape=input_shape))
model.add(tfl.MaxPooling1D())
model.add(tfl.Conv1D(64, 1, activation='relu', input_shape=input_shape))
model.add(tfl.MaxPooling1D())
model.add(tfl.Conv1D(64, 1, activation='relu', input_shape=input_shape))
model.add(tfl.MaxPooling1D())
model.add(tfl.Conv1D(64, 1, activation='relu', input_shape=input_shape))
model.add(tfl.MaxPooling1D())
model.add(tfl.Conv1D(64, 1, activation='relu', input_shape=input_shape))
model.add(tfl.MaxPooling1D())
model.add(tfl.Conv1D(64, 1, activation='relu', input_shape=input_shape))
model.add(tfl.MaxPooling1D())
model.add(tfl.Conv1D(64, 1, activation='relu', input_shape=input_shape))
model.add(tfl.MaxPooling1D())
model.add(tfl.Conv1D(64, 1, activation='relu', input_shape=input_shape))
model.add(tfl.MaxPooling1D())
model.add(tfl.Conv1D(64, 1, activation='relu', input_shape=input_shape))
model.add(tfl.MaxPooling1D())
model.add(tfl.Conv1D(64, 1, activation='relu', input_shape=input_shape))
model.compile(loss = 'sparse_categorical_crossentropy',optimizer = "adam",metrics = ['accuracy'])
model.summary()
```

```python
model.fit(x_train,y_train,batch_size=300,epochs=50,verbose=1,validation_split=0.15)
```
