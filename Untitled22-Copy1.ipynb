```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import callbacks, layers, models
from tensorflow.keras.utils import to_categorical

import numpy as np

import matplotlib.pyplot as plt
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import label_binarize
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
```

```python
df= pd.read_csv('mental-state.csv')
data_new= pd.get_dummies(df, columns = ['Label'])
```

```python
data_new
```

```python
x=data_new.loc[:,'lag1_mean_0':'freq_750_3']
x.head()
```

```python
y=data_new.loc[:,'Label_0.0':'Label_2.0']
y.head()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(x)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=50)
```

```python
X_pca = pca.fit_transform(X_scaled)
df_pca = pd.concat([pd.DataFrame(X_pca), y], axis=1)
```

```python
df_pca.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(X_pca,y,test_size=0.2,random_state=123)
x_train,x_val,y_train,y_val=train_test_split(x_train,y_train,test_size=0.2,random_state=123)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import GRU
```

```python
inputs = tf.keras.Input(shape=(x_train.shape[1],1))

gru = tf.keras.layers.GRU(256, return_sequences=True)(inputs)
flat = layers.Flatten()(gru)
outputs = layers.Dense(3, activation='softmax')(flat)

model = tf.keras.Model(inputs, outputs)

model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
#Defining a custom callback function to stop training our model when accuracy goes above 99%

class MyCallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs={}):
        if logs.get('val_acc') > 0.99:
            print("\nReached accuracy threshold! Terminating training.")
            self.model.stop_training = True
            
my_callback = MyCallback()

#EarlyStopping callback to make sure model is always learning
early_stopping = EarlyStopping(monitor='val_loss', patience=2)
```

```python
#Defining other parameters for our CNN model

model = model

METRICS = [tf.keras.metrics.CategoricalAccuracy(name='Accuracy'),
           tf.keras.metrics.AUC(name='AUC')]

CALLBACKS = [my_callback]


model.compile(optimizer='adam',
              loss=tf.losses.CategoricalCrossentropy(),
              metrics=METRICS)

model.summary()
```

```python
history=model.fit(x_train,y_train,batch_size=30,epochs=50,verbose=1,validation_data=(x_val,y_val))
```

```python
fig, ax = plt.subplots()
plt.plot(history.history['Accuracy'])

plt.plot(history.history['val_Accuracy'])

plt.title('Model Accuracy')

plt.ylabel('Accuracy')

plt.xlabel('Epochs')

plt.legend(['Training Data', 'Validation Data'], loc='lower right')

plt.show()
fig.savefig('./Results/GRU/Model_Accuracy.png', format='png', dpi=300)
```

```python
fig, ax = plt.subplots()
plt.plot(history.history['AUC'])

plt.plot(history.history['val_AUC'])

plt.title('Model AUC')

plt.ylabel('AUC')

plt.xlabel('Epochs')

plt.legend(['Training Data', 'Validation Data'], loc='lower right')

plt.show()
fig.savefig('./Results/GRU/Model_AUC.png', format='png', dpi=300)
```

```python
fig, ax = plt.subplots()
plt.plot(history.history['loss'])

plt.plot(history.history['val_loss'])

plt.title('Model Loss')

plt.ylabel('Loss')

plt.xlabel('Epochs')

plt.legend(['Training Data', 'Validation Data'], loc='upper right')

plt.show()
fig.savefig('./Results/GRU/Model_Loss.png', format='png', dpi=300)
```
