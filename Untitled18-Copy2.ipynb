```python
df= pd.read_csv('mental-state.csv')
data_new= pd.get_dummies(df, columns = ['Label'])
```

```python
x=data_new.loc[:,'lag1_mean_0':'freq_750_3']
x.head()
```

```python
y=data_new.loc[:,'Label_0.0':'Label_2.0']
y.head()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X_pca,y,test_size=0.2,random_state=123)
x_train,x_val,y_train,y_val=train_test_split(x_train,y_train,test_size=0.2,random_state=123)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
import xgboost as xgb
# XGBoost (different learning rate)
learning_rate_range = np.arange(0.01, 1, 0.07)
test_XG = [] 
train_XG = []
for lr in learning_rate_range:
    xgb_classifier = xgb.XGBClassifier(eta = lr)
    xgb_classifier.fit(x_train, y_train)
    train_XG.append(xgb_classifier.score(x_train, y_train))
    test_XG.append(xgb_classifier.score(x_test, y_test))
```

```python
fig = plt.figure(figsize=(10, 7),dpi=300)
plt.plot(learning_rate_range, train_XG, c='orange', label='Training Data')
plt.plot(learning_rate_range, test_XG, c='m', label='Testing Data')
plt.xlabel('Learning rate')
plt.xticks(learning_rate_range)
plt.ylabel('Accuracy score')
plt.ylim(0.6, 1)
plt.legend(prop={'size': 12}, loc=3)
plt.title('Accuracy Score vs. Learning Rate of XGBoost', size=14)
plt.show()
```
