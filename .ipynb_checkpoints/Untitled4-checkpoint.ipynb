```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2, mutual_info_classif
```

```python
bestfeatures = SelectKBest(score_func=mutual_info_classif, k=64)
fit = bestfeatures.fit(x,y)
```

```python
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(x.columns)
```

```python
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']
```

```python
featureScores
```

```python
print(featureScores.nlargest(64,'Score'))
```

```python
top_features=featureScores.nlargest(650,'Score')
```

```python
scaled_data =(x[top_features])
```

```python
scaled_data.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(scaled_data, y, random_state=123, test_size=0.2)
```

```python
from tensorflow.keras import Sequential
```

```python
model = Sequential()
model.add(layers.Conv1D(64, 3, input_shape=(x_train.shape[1], 1)))
model.add(layers.BatchNormalization())
model.add(layers.MaxPool1D(2))
model.add(layers.Dropout(0.5))
model.add(layers.ReLU())
    
model.add(layers.Conv1D(64, 5,activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.MaxPool1D(2))
model.add(layers.Dropout(0.5))
model.add(layers.ReLU())
    
model.add(layers.Conv1D(32, 5,activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.MaxPool1D(2))
model.add(layers.Dropout(0.5))
model.add(layers.ReLU())
model.add(layers.Flatten())
    
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dropout(0.5))
model.add(layers.Activation('relu'))
model.add(layers.Dense(3, activation='softmax'))
    
model.compile(loss = 'sparse_categorical_crossentropy',optimizer = "adam",metrics = ['accuracy'])
model.summary()
```

```python
model.fit(x_train,y_train,batch_size=300,epochs=30,verbose=1, validation_split=0.25)
```
