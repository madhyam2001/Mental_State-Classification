```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import callbacks, layers, models
from tensorflow.keras.utils import to_categorical

import numpy as np

import matplotlib.pyplot as plt
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
```

```python
df= pd.read_csv('mental-state.csv')
df.head()
```

```python
data_new= pd.get_dummies(df, columns = ['Label'])
```

```python
data_new
```

```python
# x=df.drop(['Label'],axis=1)
# y=df['Label'].values
# print(x.shape)
# print(y.shape)
```

```python
x=data_new.loc[:,'lag1_mean_0':'freq_750_3']
x.head()
```

```python
y=data_new.loc[:,'Label_0.0':'Label_2.0']
y.head()
```

```python
x=x.values
y=y.values
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=123)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
y_test
```

```python
# from sklearn.neural_network import MLPClassifier
```

```python
# base_model=MLPClassifier()
vgg_model = tf.keras.applications.vgg16.VGG16()
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import LSTM
```

```python
model = Sequential()
# model.add(base_model.layers[0])
# model.add(vgg_model.layers[0])
model.add(layers.Conv1D(64, 3, input_shape=(x_train.shape[1], 1)))
model.add(LSTM(64, return_sequences=True))
model.add(layers.BatchNormalization())
model.add(layers.MaxPool1D(2))
model.add(layers.Dropout(0.5))
model.add(layers.ReLU())


model.add(layers.Conv1D(64, 5,activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.MaxPool1D(2))
model.add(layers.Dropout(0.5))
model.add(layers.ReLU())
    
model.add(layers.Conv1D(32, 5,activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.MaxPool1D(2))
model.add(layers.Dropout(0.5))
model.add(layers.ReLU())
model.add(layers.Flatten())
model.add(layers.Dense(3, activation='softmax'))   
model.compile(loss = 'categorical_crossentropy',optimizer = "adam",metrics = ['accuracy'])
model.summary()
```

```python
model.fit(x_train,y_train,batch_size=300,epochs=30,verbose=1)
```

```python
pred = model.predict(x_test)
```
